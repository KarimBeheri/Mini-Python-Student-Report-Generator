```python
from ast import Index
# LISTS And Some Statistcs
#______________________________________________________________
names = ["  ahmed mohamed", "sara ali  ", "mohamed samir", "laila mohamed", "omar farid"]
courses = ["data analysis with python", "python for beginners", "data analysis with python", "python for beginners", "data analysis with python"]
scores = [87.4567, 92.3, 58.9, 76.45, 61.2]
scoresfinal= scores.copy()
name_cleaned = names.copy()
vip_status = [True, False, False, True, True]

for i in range(len(names)):
  name_cleaned[i]=names[i].title().strip()
  course_cleaned=courses[i].title().replace("With","with")
  score_cleaned = f"{scores[i]:.2f}"
  if(vip_status[i]==True):
    scoresfinal[i] = scores[i] + 5
  if(scoresfinal[i]>=90):
    grade="A"
  elif(89>=scoresfinal[i]>=75):
    grade="B"
  elif(74>=scoresfinal[i]>=60):
    grade="C"
  else:
    grade="F"


  print(f"Student Name: {name_cleaned}")
  print(f"Course: {course_cleaned}")
  print(f"Final Score: {scoresfinal[i]:0.2f}%")
  print(f"Grade: {grade}")

  if grade != "F":
    print("Status : Pass")
  else :
    print("Status : : Fail")


  if (vip_status[i] == True):
    print("Vip Status : Yes")
  else :
    print("Vip Status : No")


  print("-" * 30)

  # Top Score
  for j in range(len(names)):
    top_score = max(scoresfinal)
    top_score_name = name_cleaned[scoresfinal.index(top_score)]

  #Lowest Score
  for k in range(len(names)):
    top_score = min(scoresfinal)
    min_score_name = name_cleaned[scoresfinal.index(top_score)]

  #Average Score
  Average_Score = sum(scoresfinal)/len(scoresfinal)

  #Pass / Fail Counter
  pass_counter = 0
  fail_counter = 0
  for i in range(len(names)):
    if scoresfinal[i] >= 60:
      pass_counter += 1
    else:
      fail_counter += 1
  # Rank Students

sorted_indices = sorted(range(len(scoresfinal)), key=lambda x: scoresfinal[x], reverse=True)


print(f"Top Score : {max(scoresfinal)} ({top_score_name})")

print(f"Min Score : {min(scoresfinal)} ({min_score_name})")

print(f"Average Score : {Average_Score:.2f}")

print(f"Passed Student : {pass_counter}")

print(f"Failed Student : {fail_counter}")

print("\n----- Ranking -----\n")
rank =1
for  i in sorted_indices:

  print(f"{rank}. {names[i].title().strip()} - {scoresfinal[i]:.2f}")
  rank+=1






```
